```python
import cv2
import numpy as np
import time
import os
import pandas as pd
from sklearn.utils import shuffle

import psychopy.visual
import psychopy.event
```

### **These are the simulation videos :**


```python
#These are the simulation videos that are put into a dataframe (videos_df). There are a total of 20 unique videos and will eventually be shuffled the order and multiplied 5 times. 
path = '/Users/lynn/Desktop/Biological Stimulations Frame Durations/Final Videos/'
videos_df = pd.DataFrame(columns=['filename','fwd/bckwd', 'm/o', 'frame duration in ms'])
    
for v in os.listdir(path+'Forward/Original'):
    if v.endswith('.mp4'):
        videos_df =videos_df.append({'filename':'Forward/Original/'+v,'fwd/bckwd': "Forward", 'm/o': "o", 'frame duration in ms': v[10:-4]},ignore_index=True)


for v in os.listdir(path+'Forward/Mirror'):
    if v.endswith('.mp4'):
        videos_df =videos_df.append({'filename':'Forward/Mirror/'+v,'fwd/bckwd': "Forward", 'm/o': "m", 'frame duration in ms': v[12:-4]},ignore_index=True)

for v in os.listdir(path+'Reverse/Mirror'):
    if v.endswith('.mp4'):
        videos_df =videos_df.append({'filename':'Reverse/Mirror/'+v,'fwd/bckwd': "Backward", 'm/o': "o", 'frame duration in ms': v[12:-4]},ignore_index=True)
   
        
for v in os.listdir(path+'Reverse/Original'):
    if v.endswith('.mp4'):
        videos_df =videos_df.append({'filename':'Reverse/Original/' +v ,'fwd/bckwd': "Backward", 'm/o': "m", 'frame duration in ms': v[10:-4]},ignore_index=True)

videos_df
```

### ** A function to play demo videos:**


```python
path_to_demovid = '/Users/lynn/Desktop/Biological Stimulations Frame Durations/Introduction/Demo Videos/'

def play_demovid(demovid_filename):
    cap = cv2.VideoCapture(path_to_demovid + demovid_filename )
    
    if (cap.isOpened()== False): 
        print("Error opening video stream or file")
    else:
        framerate = cap.get(5)
    # Read until video is completed
    while(cap.isOpened()):
      # Capture frame-by-frame
        ret, frame = cap.read()
        if ret == True:  
            #allows the resizing and centering of the video window
            cv2.namedWindow('image', cv2.WINDOW_NORMAL)
            cv2.resizeWindow('image', 1440, 900)
            cv2.moveWindow('image', 0, 0)
            cv2.imshow('image', frame)
            cv2.waitKeyEx(40)

        else: 
            break

    
    
    #When everything done, release the video capture object
    cap.release()
    #Closes all the frames
    cv2.destroyAllWindows()
```

### ** A funtion to show images:**

```python
#This function allows the showing of an image with a specific size and location of window on screen.
path_to_img = '/Users/lynn/Desktop/Biological Stimulations Frame Durations/Introduction/Images/'

def show_image(image_filename): 
    cv2.destroyAllWindows()
    img = cv2.imread(path_to_img + image_filename)
    winname = "img"
    cv2.namedWindow(winname, cv2.WINDOW_AUTOSIZE) 
    cv2.moveWindow(winname, 0, 0)
    #cv2.resizeWindow('img', 1440,900)
    cv2.imshow(winname, img)
```

### ** A funtion to show text screen:**

```python
def show_test_image(insert_text_here, key_pressed):
    win = psychopy.visual.Window(
    size=[400, 400],
    units="pix",
    fullscr=True,
    color=[1, 1, 1])

    text = psychopy.visual.TextStim(
    win=win,
    text= insert_text_here,
    color=[-1, -1, -1])

    text.draw()

    win.flip()
    
    key_pressed = psychopy.event.waitKeys()
    
    win.close()
```

### ** The code for displaying experimental set-up:**

```python
#Enter subject name here:
subjectname = 'spacebars'

#Configuration

#Determining the 'forward' Key by pressing a key
insert_text_here = "Please press any key to indicate forward"
key_pressed = 'ForwardKey'
show_test_image(insert_text_here, key_pressed)


#Forwardkey = cv2.waitKeyEx(0)

#Determining the 'backward' Key
show_image('indicate_backward.png')
Backwardkey = cv2.waitKeyEx(0)

#Determining the 'continue' Key
show_image('indicate_continue.png')
Continuekey = cv2.waitKeyEx(0)

#show what is forward
show_image('Forward_before.png')
cv2.waitKeyEx(0)

#FORWARD DEMOVID
play_demovid('FwM13dots2sec40.mp4')
play_demovid('FwO13dots2sec40ms.mp4')

#check_forward_press
def checks_forwardkey():
    show_image('Forward_after.png')
    forward_check = cv2.waitKeyEx(0)
    if forward_check == Forwardkey:
        show_image('well_done.png')
        cv2.waitKey(500)
    else:
        show_image('try_again.png')
        cv2.waitKey(500)
        checks_forwardkey()

checks_forwardkey()
        
#show what is backward
show_image('Backward_before.png')
cv2.waitKeyEx(0)

#BACKWARD DEMOVID
play_demovid('RvO13dots2sec40ms.mp4')
play_demovid('RvM13dots2sec40.mp4')
#backward
def checks_backwardkey():
    show_image('Backward_after.png')
    backward_check = cv2.waitKeyEx(0)
    if backward_check == Backwardkey:
        show_image('well_done.png')
        cv2.waitKey(500)
    else:
        show_image('try_again.png')
        cv2.waitKey(500)
        checks_backwardkey()

checks_backwardkey()

#show what is forward
show_image('Your_task.png')
cv2.waitKeyEx(0)    

show_image('Ready.png')
cv2.waitKeyEx(0)    
        
#print 'Forward = ', Forwardkey
#print 'Backward = ', Backwardkey

data_df = pd.DataFrame()
videos_df5times = videos_df.append([videos_df, videos_df, videos_df, videos_df])
shuffled_samples = shuffle(videos_df5times
                          )

pausecounter = 25

for i, sample in shuffled_samples.iterrows():
    if pausecounter == 0:
        
        while cv2.waitKeyEx(2000) != Continuekey: #break 2 minutes
            show_image('continue.png')
        pausecounter = 25
    else:
        pausecounter -= 1
    
    shuffled_filename = sample['filename']

    cap = cv2.VideoCapture(path + shuffled_filename)

    if (cap.isOpened()== False): 
        print("Error opening video stream or file")
    else:
        framerate = cap.get(5)

    # Read until video is completed
    while(cap.isOpened()):
      # Capture frame-by-frame
        ret, frame = cap.read()
        k = -1
        if ret == True:

        #Display the resulting frame
            cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
            #cv2.resizeWindow('image', 1440, 900)
            cv2.moveWindow('image', 0, 0)
            cv2.imshow('image', frame)

        # Press Q on keyboard to  exit
            #if cv2.waitKey(0) & 0xFF == ord('8'):
                    #if (c == 2424832) x--;  //  left
            k = cv2.waitKeyEx(int(1000/framerate))
            if k == Forwardkey or k == Backwardkey:
                #k = cv2.waitKeyEx(0)
                break
            elif k != -1:
                #k = cv2.waitKeyEx(int(1000/framerate))
                time.sleep(1/framerate)

      # Break the loop
        else: 
            break

    #When everything done, release the video capture object
    cap.release()
    #Closes all the frames
    cv2.destroyAllWindows()
    if k != Forwardkey and k != Backwardkey:
        show_image('Forward_or_backward.png')
        k = cv2.waitKeyEx(0)
        
    if k == Forwardkey:
        answer = 'Forward'
    if k == Backwardkey:
        answer = 'Backward'

    if sample['fwd/bckwd'] == answer:
        sample['answer'] = 1
    else:
        sample['answer'] = 0
        
    sample['time'] =  time.ctime()
    
    data_df = data_df.append(sample)
    
show_image('all_done.png')
cv2.waitKeyEx(0)    
          
cv2.destroyAllWindows()
datapath = '/Users/lynn/Desktop/experiment data/'
data_df.to_csv(datapath+  'Data ' + subjectname +'.csv', mode = 'a')
```
